```python
import re

with open("../data/day3.txt") as f:
    data = f.read().replace("\n", "")
```

# Part 1

```python
def extract_correct_products(string):
    mul_instructions = re.findall(r"mul\((\d+,\d+)\)", string)
    ints = [tuple(map(int, x.split(","))) for x in mul_instructions]
    return ints


def sum_all_correct_products(string):
    ints = extract_correct_products(string)

    sumprod = sum(x * y for x, y in ints)

    return sumprod


len(extract_correct_products(data))

sum_all_correct_products(data)
```

```python
# The looped approach is conceptually simpler, but also a bit uglier
# I wrote this approach to check the regex variant

enabled = True
string = data
ints = []
for i in range(len(string)):
    if string[i : i + 4] == "do()":
        enabled = True
        # print(f"{enabled=}")
    if string[i : i + 7] == "don't()":
        enabled = False
        # print(f"{enabled=}")

    match = re.match(r"^mul\((\d+),(\d+)\)", string[i:])
    if match:
        int1, int2 = match.groups()
        # print(match.groups())
        if enabled:
            ints.append((int(int1), int(int2)))
sum(x * y for x, y in ints)
```
