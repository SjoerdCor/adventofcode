```python
import os
import numpy as np
```

```python
def file_to_array(filename):
    loc = os.path.join("..", "data", filename)
    with open(loc) as f:
        data = f.read()
    return np.array([list(line) for line in data.split("\n")])


data = file_to_array("day8.txt")
```

# Part 1

```python
import itertools
```

```python
def check_outside_bounds(test_position, size):
    return (
        test_position[0] < 0
        or test_position[1] < 0
        or test_position[0] >= size[0]
        or test_position[1] >= size[1]
    )


frequencies = np.unique(data[data != "."])
antinode_locations = []

for frequency in frequencies:
    locations = np.argwhere(data == frequency)
    for antenna_1, antenna_2 in itertools.combinations(locations, 2):
        diff = antenna_2 - antenna_1
        possible_antinode_location_1 = antenna_1 - diff
        if not check_outside_bounds(possible_antinode_location_1, data.shape):
            antinode_locations.append(tuple(possible_antinode_location_1))

        possible_antinode_location_2 = antenna_2 + diff
        if not check_outside_bounds(possible_antinode_location_2, data.shape):
            antinode_locations.append(tuple(possible_antinode_location_2))

len(set(antinode_locations))
```

# Part 2

```python
for frequency in frequencies:
    locations = np.argwhere(data == frequency)
    for antenna_1, antenna_2 in itertools.combinations(locations, 2):
        diff = antenna_2 - antenna_1
        i = 0
        while True:
            possible_antinode_location = antenna_1 + i * diff
            if check_outside_bounds(possible_antinode_location, data.shape):
                break
            antinode_locations.append(tuple(possible_antinode_location))
            i -= 1

        i = 1
        while True:
            possible_antinode_location = antenna_1 + i * diff
            if check_outside_bounds(possible_antinode_location, data.shape):
                break
            antinode_locations.append(tuple(possible_antinode_location))
            i += 1
```
