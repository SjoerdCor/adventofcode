```python
import numpy as np
```

```python
with open(r"../data/day4.txt") as f:
    data = f.readlines()
arr = np.array([list(line.strip()) for line in data])
```

# Part 1

```python
def find_xmas(line):
    return line.count("XMAS") + line.count("SAMX")


horizontal_matches = sum(find_xmas(line) for line in data)
print(horizontal_matches)
```

```python
vertical_matches = sum(find_xmas("".join(col)) for col in arr.T)
print(vertical_matches)
```

```python
diagonal_matches_downright = 0

for x in range(arr.shape[1]):
    for y in range(arr.shape[0]):
        if x <= (arr.shape[1] - 4) and y <= (arr.shape[0] - 4):
            word = arr[x, y] + arr[x + 1, y + 1] + arr[x + 2, y + 2] + arr[x + 3, y + 3]
            if word == "XMAS":
                diagonal_matches_downright += 1

        if x >= 3 and y >= 3:
            word = arr[x, y] + arr[x - 1, y - 1] + arr[x - 2, y - 2] + arr[x - 3, y - 3]
            if word == "XMAS":
                diagonal_matches_downright += 1

diagonal_matches_downright
```

```python
diagonal_matches_downleft = 0

for x in range(arr.shape[1]):
    for y in range(arr.shape[0]):
        if x <= (arr.shape[1] - 4) and y >= 3:
            word = arr[x, y] + arr[x + 1, y - 1] + arr[x + 2, y - 2] + arr[x + 3, y - 3]
            if word == "XMAS":
                diagonal_matches_downleft += 1

        if x >= 3 and y <= (arr.shape[0] - 4):
            word = arr[x, y] + arr[x - 1, y + 1] + arr[x - 2, y + 2] + arr[x - 3, y + 3]
            if word == "XMAS":
                diagonal_matches_downleft += 1

diagonal_matches_downleft
```

```python
diagonal_matches_downleft = 0
arr_rev = np.fliplr(arr)
for horizontal_offset in range(arr_rev.shape[1]):
    diag = "".join(np.diag(arr_rev, horizontal_offset))
    diagonal_matches_downleft += find_xmas(diag)
for vertical_offset in range(arr_rev.shape[0]):
    diag = "".join(np.diag(arr_rev[vertical_offset:], 0))
    diagonal_matches_downleft += find_xmas(diag)

diagonal_matches_downleft
```

```python
(
    horizontal_matches
    + vertical_matches
    + diagonal_matches_downleft
    + diagonal_matches_downright
)
```

# Part 2
Find a cross of mas

```python
def check_3x3(arr):
    downright = (arr[0, 0] + arr[1, 1] + arr[2, 2]) in ("MAS", "SAM")
    downleft = (arr[0, 2] + arr[1, 1] + arr[2, 0]) in ("MAS", "SAM")
    return downright and downleft
```

```python
number_xmases = 0
for x in range(arr.shape[1] - 2):
    for y in range(arr.shape[0] - 2):
        temp_arr = arr[x : x + 3, y : y + 3]
        number_xmases += check_3x3(temp_arr)
```
