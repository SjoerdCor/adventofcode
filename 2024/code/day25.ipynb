```python
import numpy as np

with open("../data/day25.txt") as f:
    data = f.read()

sections = data.split("\n\n")


def string_to_array(string):
    return np.array([list(row) for row in string.split("\n")])


locks = []
keys = []
for section in sections:
    arr = string_to_array(section)
    if section[0] == "#":
        heights = np.argmax(arr[1:] == ".", axis=0)
        locks.append(heights)
    if section[0] == ".":
        heights = 5 - np.argmax(arr[1:] == "#", axis=0)
        keys.append(heights)
```

```python
fitting_pairs = 0
for lock in locks:
    for key in keys:
        fitting_pairs += ((lock + key) < 6).all()
fitting_pairs
```
