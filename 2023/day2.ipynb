```python
with open("day2.txt") as f:
    data = f.read()
data = data.splitlines()
```

```python
def parse_subset(subset):
    dct = {}
    for combi in subset.split(","):
        n, color = combi.split()
        dct[color] = int(n)
    return dct


def parse_line(line):
    game, cubesubsets = line.split(":")
    id_ = int(game[5:])

    subsets = cubesubsets.split(";")
    result = [parse_subset(subset) for subset in subsets]
    return id_, result


def find_all_games(data):
    games = {}
    for line in data:
        id_, result = parse_line(line)
        games[id_] = result
    return games


def subset_is_valid(subset):
    MAX_VALUES = {"red": 12, "green": 13, "blue": 14}
    for item, max_n in MAX_VALUES.items():
        if subset.get(item, 0) > max_n:
            return False
    return True


games = find_all_games(data)

valid_games = [
    id_
    for id_, subsets in games.items()
    if all(subset_is_valid(subset) for subset in subsets)
]
sum(valid_games)
```

```python
def find_needed(subsets, color):
    return max(subset.get(color, 0) for subset in subsets)


def find_power(subsets):
    colors = ["blue", "red", "green"]
    ans = 1
    for color in colors:
        ans *= find_needed(subsets, color)
    return ans


ans = 0
for subsets in games.values():
    ans += find_power(subsets)
ans
```
