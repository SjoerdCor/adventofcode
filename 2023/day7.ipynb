```python
import aoc

data = aoc.read("day7.txt").splitlines()
data = [(line.split()[0], int(line.split()[1])) for line in data]
```

```python
from collections import Counter
```

```python
def get_combo_rank(cntr):
    vals = cntr.values()
    if 5 in vals:
        return 1  # Five of a kind
    if 4 in vals:
        return 2  # Four of a kind
    if 3 in vals:
        if 2 in vals:
            return 3  # Full house
        return 4  # 3 of a kind
    if 2 in vals:
        if Counter(vals)[2] == 2:
            return 5  # 2 pair
        return 6  # single pair
    return 7  # high card


def get_card_rank(card):
    cards = "AKQJT98765432"
    return cards.index(card)


def get_hand_rank(hand):
    combos = Counter(hand)
    return (
        get_combo_rank(combos),
        get_card_rank(hand[0]),
        get_card_rank(hand[1]),
        get_card_rank(hand[2]),
        get_card_rank(hand[3]),
        get_card_rank(hand[4]),
    )


data.sort(key=lambda x: get_hand_rank(x[0]), reverse=True)
ans = 0
for rank, (hand, bid) in enumerate(data, start=1):
    ans += rank * bid
print(ans)
```

# Part 2

```python
def get_combo_rank(cntr):
    vals = cntr.values()
    if 5 in vals:
        return "1FiveOfAKind"  # Five of a kind
    if 4 in vals:
        return "2FourOfAKind"  # Four of a kind
    if 3 in vals:
        if 2 in vals:
            return "3FullHouse"  # Full house
        return "4ThreeOfAKind"  # 3 of a kind
    if 2 in vals:
        if Counter(vals)[2] == 2:
            return "5TwoPair"  # 2 pair
        return "6SinglePair"  # single pair
    return "7HighCard"  # high card


def get_card_rank_with_joker(card):
    cards = "AKQT98765432J"
    try:
        return cards.index(card)
    except ValueError:
        print(card)


def replace_jokers_in_hand(cntr):
    max_val = None
    max_occ = 0

    for val, occ in cntr.items():
        if val == "J":
            continue
        if occ > max_occ:
            max_val = val
            max_occ = occ

    cntr[max_val] += cntr["J"]
    del cntr["J"]

    return cntr


def get_hand_rank_with_joker(hand):
    combos = Counter(hand)
    combos = replace_jokers_in_hand(combos)

    return (
        get_combo_rank(combos),
        get_card_rank_with_joker(hand[0]),
        get_card_rank_with_joker(hand[1]),
        get_card_rank_with_joker(hand[2]),
        get_card_rank_with_joker(hand[3]),
        get_card_rank_with_joker(hand[4]),
    )


test_data = data[::1]
test_data.sort(key=lambda x: get_hand_rank_with_joker(x[0]), reverse=True)
ans = 0
for rank, (hand, bid) in enumerate(test_data, start=1):
    ans += rank * bid
print(ans)
```
