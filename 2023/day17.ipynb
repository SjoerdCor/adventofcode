```python
import aoc

data = aoc.read("day17.txt", to_grid=True)
rows, cols = len(data), len(data[0])
```

# Part 1

```python
heap = [(0, (0, 0), "X", 0)]
MAX_STEPS_SAME_DIRECTION = 3

import heapq


def is_legal_move(new_pos, current_dir, n_times_dir, new_direction):
    if aoc.is_outside_bounds(new_pos, (rows, cols)):
        return False

    # Cannot reverse
    if current_dir + new_direction in {"SN", "NS", "WE", "EW"}:
        return False

    if new_dir_name == current_direction and n_times_dir == MAX_STEPS_SAME_DIRECTION:
        return False
    return True


seen = {}
while heap:
    heatloss, current_pos, current_direction, n_steps_dir = heapq.heappop(heap)
    for new_dir_name, new_dir_value in aoc.DIRECTIONS.items():
        new_pos = tuple(current_pos + new_dir_value)
        same_dir = new_dir_name == current_direction
        legal = is_legal_move(new_pos, current_direction, n_steps_dir, new_dir_name)
        if not legal:
            continue
        r, c = new_pos
        new_heatloss = heatloss + int(data[r][c])

        if same_dir:
            new_n_steps_dir = n_steps_dir + 1
        else:
            new_n_steps_dir = 1

        for steps_for_dict in range(new_n_steps_dir, MAX_STEPS_SAME_DIRECTION + 1):

            state = (new_pos, new_dir_name, steps_for_dict)
            if new_heatloss < seen.get(state, float("inf")):
                seen[state] = new_heatloss
                if steps_for_dict == new_n_steps_dir:
                    full_heap_state = (
                        new_heatloss,
                        new_pos,
                        new_dir_name,
                        new_n_steps_dir,
                    )
                    if full_heap_state not in heap:
                        heapq.heappush(heap, full_heap_state)

        if new_pos == (rows - 1, cols - 1):
            print(new_heatloss)


min(value for key, value in seen.items() if key[0] == (rows - 1, cols - 1))
```

# Part 2

```python
MIN_STEPS_SAME_DIRECTION_ULTRA = 4
MAX_STEPS_SAME_DIRECTION_ULTRA = 10

import heapq


def is_legal_move_ultra(new_pos, current_dir, n_times_dir, new_direction):
    if aoc.is_outside_bounds(new_pos, (rows, cols)):
        return False

    # Cannot reverse
    if current_dir + new_direction in {"SN", "NS", "WE", "EW"}:
        return False

    if (
        new_dir_name == current_direction
        and n_times_dir == MAX_STEPS_SAME_DIRECTION_ULTRA
    ):
        return False
    if (
        new_dir_name != current_direction
        and n_times_dir < MIN_STEPS_SAME_DIRECTION_ULTRA
    ):
        return False

    return True


heap = [(0, (0, 0), "E", 0), (0, (0, 0), "S", 0)]
seen = {}
while heap:
    # print(heap)
    heatloss, current_pos, current_direction, n_steps_dir = heapq.heappop(heap)
    for new_dir_name, new_dir_value in aoc.DIRECTIONS.items():
        new_pos = tuple(current_pos + new_dir_value)
        legal = is_legal_move_ultra(
            new_pos, current_direction, n_steps_dir, new_dir_name
        )
        if not legal:
            continue
        r, c = new_pos
        new_heatloss = heatloss + int(data[r][c])

        if new_dir_name == current_direction:
            new_n_steps_dir = n_steps_dir + 1
        else:
            new_n_steps_dir = 1

        state = (new_pos, new_dir_name, new_n_steps_dir)
        if new_heatloss < seen.get(state, float("inf")):
            seen[state] = new_heatloss
            full_heap_state = (new_heatloss, new_pos, new_dir_name, new_n_steps_dir)
            if full_heap_state not in heap:
                heapq.heappush(heap, full_heap_state)

        if new_pos == (rows - 1, cols - 1):
            print(new_heatloss)


min(
    value
    for key, value in seen.items()
    if key[0] == (rows - 1, cols - 1) and key[-1] >= MIN_STEPS_SAME_DIRECTION_ULTRA
)
```
