```python
import aoc

data = aoc.read("day17.txt", to_grid=True)
rows, cols = len(data), len(data[0])
```

# Part 1

```python
import heapq
```

```python
MAX_STEPS_SAME_DIRECTION = 3


def is_legal_move(new_pos, current_dir, n_times_dir, new_direction):
    if aoc.is_outside_bounds(new_pos, (rows, cols)):
        return False

    # Cannot reverse
    if current_dir + new_direction in {"SN", "NS", "WE", "EW"}:
        return False

    if new_dir_name == current_direction and n_times_dir == MAX_STEPS_SAME_DIRECTION:
        return False
    return True


heap = [(0, (0, 0), "X", 0)]
seen = set()
while heap:
    heatloss, current_pos, current_direction, n_steps_dir = heapq.heappop(heap)
    if current_pos == (rows - 1, cols - 1):
        print(heatloss)
        break
    state = (current_pos, current_direction, n_steps_dir)
    if state in seen:
        continue
    seen.add(state)
    for new_dir_name, new_dir_value in aoc.DIRECTIONS.items():
        new_pos = tuple(current_pos + new_dir_value)
        legal = is_legal_move(new_pos, current_direction, n_steps_dir, new_dir_name)
        if not legal:
            continue

        r, c = new_pos
        new_heatloss = heatloss + int(data[r][c])

        if new_dir_name == current_direction:
            new_n_steps_dir = n_steps_dir + 1
        else:
            new_n_steps_dir = 1
        heapq.heappush(heap, (new_heatloss, new_pos, new_dir_name, new_n_steps_dir))
```

# Part 2

```python
MIN_STEPS_SAME_DIRECTION_ULTRA = 4
MAX_STEPS_SAME_DIRECTION_ULTRA = 10


def is_legal_move_ultra(new_pos, current_dir, n_times_dir, new_direction):
    if aoc.is_outside_bounds(new_pos, (rows, cols)):
        return False

    # Cannot reverse
    if current_dir + new_direction in {"SN", "NS", "WE", "EW"}:
        return False

    if (
        new_dir_name == current_direction
        and n_times_dir == MAX_STEPS_SAME_DIRECTION_ULTRA
    ):
        return False
    if (
        new_dir_name != current_direction
        and n_times_dir < MIN_STEPS_SAME_DIRECTION_ULTRA
    ):
        return False

    return True


heap = [(0, (0, 0), "E", 0), (0, (0, 0), "S", 0)]
seen = set()
while heap:
    heatloss, current_pos, current_direction, n_steps_dir = heapq.heappop(heap)
    if (
        current_pos == (rows - 1, cols - 1)
        and n_steps_dir >= MIN_STEPS_SAME_DIRECTION_ULTRA
    ):
        print(heatloss)
        break
    state = (current_pos, current_direction, n_steps_dir)
    if state in seen:
        continue
    seen.add(state)
    for new_dir_name, new_dir_value in aoc.DIRECTIONS.items():
        new_pos = tuple(current_pos + new_dir_value)
        legal = is_legal_move_ultra(
            new_pos, current_direction, n_steps_dir, new_dir_name
        )
        if not legal:
            continue
        r, c = new_pos
        new_heatloss = heatloss + int(data[r][c])

        if new_dir_name == current_direction:
            new_n_steps_dir = n_steps_dir + 1
        else:
            new_n_steps_dir = 1

        new_state = (new_heatloss, new_pos, new_dir_name, new_n_steps_dir)
        heapq.heappush(heap, new_state)
```
