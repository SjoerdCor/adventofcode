```python
import aoc

data = aoc.read("day4.txt").splitlines()
```

```python
def parse_line(line):
    game, scratchcard = line.split(":")
    id_ = int(game[5:])
    my_numbers, winning_numbers = scratchcard.split("|")

    my_numbers = aoc.to_ints(my_numbers.split())
    winning_numbers = set(aoc.to_ints(winning_numbers.split()))
    return id_, (my_numbers, winning_numbers)


def find_all_games(data):
    games = {}
    for line in data:
        id_, result = parse_line(line)
        games[id_] = result
    return games


games = find_all_games(data)
```

```python
def score_game(scratchcard):
    my_numbers, winning_numbers = scratchcard
    matching_numbers = sum(num in winning_numbers for num in my_numbers)
    return int(2 ** (matching_numbers - 1))


sum(score_game(g) for g in games.values())
```

```python
def n_matches(scratchcard):
    my_numbers, winning_numbers = scratchcard
    return sum(num in winning_numbers for num in my_numbers)
```

```python
n_cards = {i: 1 for i in games.keys()}

for i, n in n_cards.items():
    matches = n_matches(games[i])

    for j in range(1, matches + 1):
        n_cards[i + j] += n_cards[i]
```
