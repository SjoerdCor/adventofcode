```python
import aoc

data = aoc.read("day6.txt").splitlines()
```

```python
times = aoc.to_ints(data[0].split(":")[1].split())
distances = aoc.to_ints(data[1].split(":")[1].split())
```

```python
def get_ways_to_win(time, distance):
    if time**2 / 4 < distance:
        return 0

    max_ = time // 2 + 1
    for i in range(max_):
        if (time - i) * i > distance:
            is_even = time % 2 == 0
            return (max_ - i) * 2 - is_even


ans = 1
for t, d in zip(times, distances):
    ans *= get_ways_to_win(t, d)
ans
```

# Part 2

```python
time = int("".join(str(t) for t in times))
distance = int("".join(str(d) for d in distances))


def wins_race(held_time, total_time, distance):
    return held_time * (total_time - held_time) > distance


max_ = time // 2 + 1
first_held_time = aoc.find_first_true(
    lambda i: wins_race(i, time, distance), range(max_)
)
is_even = time % 2 == 0
print((max_ - first_held_time) * 2 - is_even)
```
