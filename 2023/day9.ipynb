```python
import aoc

data = aoc.read("day9.txt")
```

```python
data = [aoc.to_ints(line.split()) for line in data.splitlines()]
```

# Part 1

```python
def calc_diff(list):
    return [list[i + 1] - list[i] for i in range(len(list) - 1)]


def extrapolate(line):

    additions = []

    diffs = calc_diff(line)

    additions.append(diffs[-1])


    while len(set(diffs)) > 1:

        diffs = calc_diff(diffs)

        additions.append(diffs[-1])

    return line[-1] + sum(additions)



sum([extrapolate(line) for line in data])
```

# Part 2

```python
def extrapolate_backwards(line):
    additions = []
    diffs = calc_diff(line)
    additions.append(diffs[0])
    while len(set(diffs)) > 1:
        diffs = calc_diff(diffs)
        additions.append(diffs[0])

    cv = additions[-1]
    additions = additions[::-1]
    for a in additions[1:]:
        cv = a - cv
    return line[0] - cv


sum([extrapolate_backwards(line) for line in data])
```
