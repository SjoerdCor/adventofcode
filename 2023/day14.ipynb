# Part 1

```python
import copy


def tilt_north(data):
    block_copy = copy.deepcopy(data)

    rows = len(data)
    cols = len(data[0])
    r = 1
    while True:
        r = max(1, r)
        moved_away = False
        for c in range(cols):
            if block_copy[r][c] == "O" and block_copy[r - 1][c] == ".":
                block_copy[r - 1][c] = "O"
                block_copy[r][c] = "."
                moved_away = True
        if moved_away:
            r -= 1
        else:
            if r == rows - 1:
                return block_copy
            r += 1


tilted = tilt_north(data)


def calc_load_on_north_support_beams(platform):
    rows = len(platform)

    load = 0
    for r, row in enumerate(platform):
        for c in row:
            if c == "O":
                load += rows - r

    return load


calc_load_on_north_support_beams(tilted)
```

# Part 2

```python
import functools
```

```python
def list_of_lists_to_tuple_of_tuples(lst_of_lsts):
    return tuple(tuple(line) for line in lst_of_lsts)


def tuple_of_tuple_to_list_of_lists(tup_of_tups):
    return list(list(line) for line in tup_of_tups)


@functools.cache
def tilt_north(data):
    block_copy = copy.deepcopy(tuple_of_tuple_to_list_of_lists(data))

    rows = len(data)
    cols = len(data[0])
    r = 1
    while True:
        r = max(1, r)
        moved_away = False
        for c in range(cols):
            if block_copy[r][c] == "O" and block_copy[r - 1][c] == ".":
                block_copy[r - 1][c] = "O"
                block_copy[r][c] = "."
                moved_away = True
        if moved_away:
            r -= 1
        else:
            if r == rows - 1:
                return list_of_lists_to_tuple_of_tuples(block_copy)
            r += 1


@functools.cache
def tilt_south(data):
    block_copy = copy.deepcopy(tuple_of_tuple_to_list_of_lists(data))

    rows = len(data)
    cols = len(data[0])
    r = rows - 2
    while True:
        r = min(rows - 2, r)
        moved_away = False
        for c in range(cols):
            if block_copy[r][c] == "O" and block_copy[r + 1][c] == ".":
                block_copy[r + 1][c] = "O"
                block_copy[r][c] = "."
                moved_away = True
        if moved_away:
            r += 1
        else:
            if r == 0:
                return list_of_lists_to_tuple_of_tuples(block_copy)
            r -= 1


@functools.cache
def tilt_east(data):
    block_copy = copy.deepcopy(tuple_of_tuple_to_list_of_lists(data))

    rows = len(data)
    cols = len(data[0])
    c = cols - 2
    while True:
        c = min(cols - 2, c)
        moved_away = False
        for r in range(rows):
            if block_copy[r][c] == "O" and block_copy[r][c + 1] == ".":
                block_copy[r][c + 1] = "O"
                block_copy[r][c] = "."
                moved_away = True
        if moved_away:
            c += 1
        else:
            if c == 0:
                return list_of_lists_to_tuple_of_tuples(block_copy)
            c -= 1


@functools.cache
def tilt_west(data):
    block_copy = copy.deepcopy(tuple_of_tuple_to_list_of_lists(data))

    rows = len(data)
    cols = len(data[0])
    c = 1
    while True:
        c = max(1, c)
        moved_away = False
        for r in range(rows):
            if block_copy[r][c] == "O" and block_copy[r][c - 1] == ".":
                block_copy[r][c - 1] = "O"
                block_copy[r][c] = "."
                moved_away = True
        if moved_away:
            c -= 1
        else:
            if c == cols - 1:
                return list_of_lists_to_tuple_of_tuples(block_copy)
            c += 1


@functools.cache
def cycle_tilt(platform):
    platform = tilt_north(platform)
    platform = tilt_west(platform)
    platform = tilt_south(platform)
    platform = tilt_east(platform)
    return platform


def find_cycle_of_cycles(platform):
    reached_states = [platform]
    reached_states_set = set(platform)
    while True:
        platform = cycle_tilt(platform)

        if platform in reached_states_set:
            first_time_state = reached_states.index(platform)
            length_cycle = len(reached_states) - first_time_state
            return first_time_state, length_cycle
        reached_states.append(platform)
        reached_states_set.add(platform)


platform_tup = list_of_lists_to_tuple_of_tuples(data)
first_time_state, length_cycle = find_cycle_of_cycles(platform_tup)
```

```python
def cycle_many_tilts(platform, n=1_000_000_000):
    for _ in range(n):
        platform = cycle_tilt(platform)
    return platform


n = 1_000_000_000
place_in_cycle = (n - first_time_state) % length_cycle
final_platform = cycle_many_tilts(platform_tup, n=first_time_state + place_in_cycle)
calc_load_on_north_support_beams(final_platform)
```
