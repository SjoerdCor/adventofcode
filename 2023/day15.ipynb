# Part 1

```python
import functools
```

```python
@functools.cache
def HASH(string):
    value = 0
    for c in string:
        value += ord(c)
        value *= 17
        value %= 256
    return value


sum(HASH(sub) for sub in data.split(","))
```

# Part 2

```python
steps = data.split(",")
```

```python
lens_slots = {i: [] for i in range(256)}


def find_label_in_box(sublabel, box):
    current_place = None
    for j, place in enumerate(box):
        if place[0] == sublabel:
            current_place = j
    return current_place


def parse_step(step):
    if step[-2] == "=":
        operation = "="
        focal_length = int(step[-1])
        label = step[:-2]
    elif step.endswith("-"):
        operation = "-"
        focal_length = None
        label = step[:-1]
    else:
        raise RuntimeError(f"Unknown operation for {step=}")
    return label, operation, focal_length


def execute_operation(box, sublabel, operation, focal_length, current_place):
    if operation == "=":
        if current_place is None:
            box.append((sublabel, focal_length))
        else:
            box[current_place] = (sublabel, focal_length)
    elif operation == "-":
        if current_place is not None:
            del box[current_place]
    return box


def handle_step(step, lens_slots):
    label, operation, focal_length = parse_step(step)

    hash_code = HASH(label)
    current_place = find_label_in_box(label, lens_slots[hash_code])

    box = execute_operation(
        lens_slots[hash_code], label, operation, focal_length, current_place
    )
    lens_slots[hash_code] = box
    return lens_slots


for step in steps:
    lens_slots = handle_step(step, lens_slots)
```

```python
def score_lens_slots(lens_slots):
    ans = 0
    for i, filling in lens_slots.items():
        for j, filler in enumerate(filling, start=1):
            ans += (i + 1) * j * filler[1]
    return ans


score_lens_slots(lens_slots)
```
