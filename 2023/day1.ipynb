```python
with open("day1.txt") as f:
    data = f.read()

lines = data.splitlines()
```

```python
def extract_first_number(string):
    for c in string:
        if c.isdigit():
            return int(c)
    raise ValueError("Could not find number")


def find_two_digit_number(string):
    forward_number = extract_first_number(string)
    backward_number = extract_first_number(string[::-1])
    return int(str(forward_number) + str(backward_number))


sum(find_two_digit_number(row) for row in lines)
```

# Part 2

```python
def extract_first_number_including_text(string, direction=1):
    if direction == -1:
        start, stop = len(string) - 1, -1
    if direction == 1:
        start, stop = 0, len(string)
    textnumbers = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9,
    }
    for i in range(start, stop, direction):
        c = string[i]
        if c.isdigit():
            return int(c)
        for textnumber, value in textnumbers.items():
            if string[i:].startswith(textnumber):
                return value
    raise ValueError("Could not find number")


def find_two_digit_number(string):
    forward_number = extract_first_number_including_text(string)
    backward_number = extract_first_number_including_text(string, direction=-1)
    return int(str(forward_number) + str(backward_number))


sum(find_two_digit_number(row) for row in lines)
```
