```python
import aoc

data = aoc.read("day16.txt", to_grid=True)
```

# Part 1

```python
from collections import deque

# 0 : north,


def find_next_directions(char, direction):
    if char == ".":
        return direction
    if char == "-":
        if direction in "WE":
            return direction
        return "WE"
    if char == "|":
        if direction in "NS":
            return direction
        return "NS"
    if char == r"/":
        if direction == "N":
            return "E"
        if direction == "W":
            return "S"
        if direction == "S":
            return "W"
        if direction == "E":
            return "N"
    if char == "\\":
        if direction == "N":
            return "W"
        if direction == "E":
            return "S"
        if direction == "S":
            return "E"
        if direction == "W":
            return "N"


rows = len(data)
cols = len(data[0])

streaming_beam = [((0, 0), "E")]
found_states = set()
while streaming_beam:
    current_spot = streaming_beam.pop()
    current_pos, current_direction = current_spot
    if (
        aoc.is_outside_bounds(current_spot[0], (rows, cols))
        or current_spot in found_states
    ):
        continue
    found_states.add(current_spot)

    r, c = current_pos
    new_directions = find_next_directions(data[r][c], current_direction)
    for new_dir in new_directions:
        streaming_beam.append((tuple(current_pos + aoc.DIRECTIONS[new_dir]), new_dir))

len(set(state[0] for state in found_states))
```

# Part 2

```python
def find_n_energized_tiles(streaming_beam):
    found_states = set()
    while streaming_beam:
        current_spot = streaming_beam.pop()
        current_pos, current_direction = current_spot
        if (
            aoc.is_outside_bounds(current_spot[0], (rows, cols))
            or current_spot in found_states
        ):
            continue
        found_states.add(current_spot)

        r, c = current_pos
        new_directions = find_next_directions(data[r][c], current_direction)
        for new_dir in new_directions:
            streaming_beam.append(
                (tuple(current_pos + aoc.DIRECTIONS[new_dir]), new_dir)
            )
    return len(set(state[0] for state in found_states))


all_starting_positions_top = [((0, c), "S") for c in range(cols)]
all_starting_positions_bottom = [((rows - 1, c), "N") for c in range(cols)]
all_starting_positions_left = [((r, 0), "E") for r in range(rows)]
all_starting_positions_right = [((r, cols - 1), "W") for r in range(rows)]

all_starting_positions = (
    all_starting_positions_top
    + all_starting_positions_left
    + all_starting_positions_bottom
    + all_starting_positions_right
)


results = {}
for starting_position in all_starting_positions:
    streaming_beam = [starting_position]
    results[starting_position] = find_n_energized_tiles(streaming_beam)
max(results.values())
```
