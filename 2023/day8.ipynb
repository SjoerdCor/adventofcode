```python
import aoc

data = aoc.read("day8.txt")
instruction, network_str = data[0], data[1].splitlines()
```

```python
import re
```

```python
def parse_network(network_str):
    network = {}
    for bla in network_str:
        node, edges = bla.split("=")
        edges = re.match(r"\(([0-9A-Z]{3}), ([0-9A-Z]{3})\)", edges.strip()).groups()
        network[node.strip()] = edges
    return network


network = parse_network(network_str)
```

# Part 1

```python
def count_steps(network):
    begin_location = "AAA"
    end_location = "ZZZ"

    location = begin_location
    i = 0
    while location != end_location:
        direction = instruction[i % len(instruction)]
        to_choose = 0 if direction == "L" else 1
        location = network[location][to_choose]
        i += 1
    return i


count_steps(network)
```

# Part 2

Calculate all the steps to ending nodes and then find common multiples

```python
def find_z_locations(visited_states, current_state):
    start_loop = visited_states.index(current_state)
    loop_length = len(visited_states) - start_loop
    z_locations = [i for i, state in enumerate(visited_states) if state[0][-1] == "Z"]
    return z_locations, loop_length


def find_looped_locations(begin_location):
    location = begin_location

    # We need both a set and a list; the set is MUCH faster to look up whether we
    # are already in a loop. But we need the list for the following step to find
    # the startingpoint and length of the loop
    visited_states = []
    visited_states_set = set()
    for i in range(1_000_000_000):
        i_mod = i % len(instruction)
        state = (location, i_mod)
        if state in visited_states_set:
            return find_z_locations(visited_states, state)
        visited_states.append(state)
        visited_states_set.add(state)
        direction = instruction[i_mod]
        to_choose = 0 if direction == "L" else 1
        location = network[location][to_choose]
    raise RuntimeError("Could not find loop!")


begin_locations = [node for node in network.keys() if node[-1] == "A"]
looped_locations = {loc: find_looped_locations(loc) for loc in begin_locations}
```

```python
looped_locations  # "Luckily", the loop only contains 1 node that ends with "Z", and it is always at exactly the index of the length
# Therefore we can just find the lowest common multiple
```
