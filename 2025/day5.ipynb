```python
import aoc

input_ = aoc.read("day5.txt")

fresh_ranges = []
for r in input_[0].splitlines():
    a, b = r.split("-")
    fresh_ranges.append((int(a), int(b)))

available_ingredients = aoc.to_ints(input_[1].splitlines())
```

```python
def is_fresh(ingredient, fresh_ranges):
    for r_min, r_max in fresh_ranges:
        if r_min <= ingredient <= r_max:
            return True
    return False

fresh_ingredients = [ing for ing in available_ingredients if is_fresh(ing, fresh_ranges)]
print(f"Number of fresh ingredients: {len(fresh_ingredients)}")
```

```python
def combine_ranges(r1, r2):
    if r2[0] < r1[0]:
        r1, r2 = r2, r1
    
    if r1[1] < r2[0]:
        return set([r1, r2])
    if r1[1] > r2[1]:
        return set([r1])
    return set([(r1[0], r2[1])])

assert combine_ranges((2, 3), (6, 7)) == {(2, 3), (6, 7)}
assert combine_ranges((2, 6), (3, 4)) == {(2, 6)}
assert combine_ranges((2, 6), (4, 8)) == {(2, 8)}
assert combine_ranges((6, 7), (2, 3)) == {(2, 3), (6, 7)}
```

```python
def add_new_range(new_range, current_ranges):
    new_ranges = set()
    for current_range in current_ranges:
        combined = combine_ranges(new_range, current_range)
        if len(combined) == 2:
            new_ranges.update([current_range])
        else:
            new_range = list(combined)[0]
            current_ranges.remove(current_range)
            return add_new_range(new_range, current_ranges)
    new_ranges.update([new_range])
    return new_ranges    

assert add_new_range((2,4), {(6, 8)}) == {(2, 4), (6, 8)}
assert add_new_range((2,4), {(3, 8)}) == {(2, 8)}
assert add_new_range((2,4), {(3, 8), (12, 16)}) == {(2, 8), (12, 16)}
```

```python
all_fresh = {fresh_ranges[0]}
for r in fresh_ranges[1:]:
    all_fresh = add_new_range(r, all_fresh)

sum(r[1] - r[0] + 1 for r in all_fresh)
```
