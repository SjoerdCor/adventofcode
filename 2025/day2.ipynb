```python
import aoc
input = aoc.read("day2.txt").split(',')

ranges = []
for x in input:
    parts = x.split('-')
    start = int(parts[0])
    end = int(parts[1]) + 1
    ranges.append(range(start, end))
```

```python
def is_invalid_id(id_):
    s = str(id_)
    l = len(s)
    if l % 2 == 1:
        return False
    h = l // 2

    return s[:h] == s[h:]

ans = 0
for range_ in ranges:
    for id_ in range_:
        if is_invalid_id(id_):
            ans += id_
print(ans)
```

```python

def check_invalid_length(s, length):
    pattern = s[:length]
    l = len(s)
    for i in range(length, l, length):
        if s[i: i + length] != pattern:
            return False
    return True


def is_invalid_id(id_):
    s = str(id_)
    l = len(s)
    for length in range(1, l//2 + 1):
        if l % length != 0:
            continue
        invalid = check_invalid_length(s, length)
        if invalid:
            return True
    return False

assert check_invalid_length("1515", 2)
assert is_invalid_id(1515)
```

```python
ans = 0
for range_ in ranges:
    for id_ in range_:
        if is_invalid_id(id_):
            ans += id_
print(ans)
```
