```python
import aoc
```

```python
password = aoc.read("day1.txt").splitlines()
```

```python
def move(position, instruction):
    original_position = position  
    direction = instruction[0]
    nr = int(instruction[1:])
    rounds, change = divmod(nr, 100)
    if direction == "L":
        position -= change
    elif direction == "R":
        position += change
    else:
        raise ValueError("Invalid instruction")
    
    passed_0 = (((position <= 0) & (original_position != 0)) | (position >= 100))
    return position % 100, passed_0 + rounds
```

```python
position = 50 # Start
times_at_0 = 0
times_passing_0 = 0
for instruction in password:
    position, passed_0 = move(position, instruction)
    times_passing_0 += passed_0
    if position == 0:
        times_at_0 += 1

print(times_at_0)
print(times_passing_0)
```
