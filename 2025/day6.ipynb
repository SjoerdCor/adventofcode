```python
import aoc
import numpy as np


homework = aoc.read("day6.txt").split('\n')
operators = homework.pop().split()
numbers = np.transpose([aoc.to_ints(numbers.split()) for numbers in homework])
```

```python
ans = 0
for prob_numbers, operator in zip(numbers, operators):
    if operator == "*":
        ans += np.prod(prob_numbers)
    elif operator == "+":
        ans += np.sum(prob_numbers) 
    else:
        raise ValueError()

ans
```

# Part 2

```python
problems = []
problem = []
for vertical_nrs in zip(*homework):
    number = ''
    for l in vertical_nrs:
        if l != ' ':
            number += l
    if not number:
        problems.append(problem)
        problem = []
    else: 
        problem.append(int(number))

problems.append(problem)
```

```python
ans = 0
for prob_numbers, operator in zip(problems, operators):
    if operator == "*":
        ans += np.prod(prob_numbers)
    elif operator == "+":
        ans += np.sum(prob_numbers) 
    else:
        raise ValueError()

ans
```
