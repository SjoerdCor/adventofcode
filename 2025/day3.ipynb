```python
import aoc
input = aoc.read("day3.txt").splitlines()
```

# Part 1

```python
def find_largest_joltage(bank):
    max_first_nr = max(bank[:-1])
    idx_first_nr = bank.index(max_first_nr)
    second_nr = max(bank[idx_first_nr + 1:])
    return int(max_first_nr + second_nr)


ans = 0
for bank in input:
    ans += find_largest_joltage(bank)

print(ans)
```

# Part 2

```python
def find_largest_joltage_of_length(bank, length, current=''):
    usable_bank = bank[:-length + 1] if length > 1 else bank
    max_nr = max(usable_bank)
    if length == 1:
        return int(current + max_nr)
    idx = bank.index(max_nr)
    return find_largest_joltage_of_length(bank[idx + 1:], length - 1, current + str(max_nr))

ans = 0
for bank in input:
    ans += find_largest_joltage_of_length(bank, 12)

print(ans)
```
