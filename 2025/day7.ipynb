```python
import aoc
```

```python
diagram = aoc.read("day7.txt", to_grid=True)
s = (len(diagram), len(diagram[0]))
```

```python
n_splits = 0
beam_positions = {(0, 70)}
while beam_positions:
    new_positions = set()
    for p in beam_positions:
        candidate_position = (p[0] + 1, p[1])
        if aoc.is_outside_bounds(candidate_position, s):
            continue
        if diagram[candidate_position[0]][candidate_position[1]] == "^":
            n_splits += 1
            new_positions.update([(p[0] + 1, p[1] - 1)])
            new_positions.update([(p[0] + 1, p[1] + 1)])
        else: 
            new_positions.update([candidate_position])
    beam_positions = new_positions.copy()

n_splits
```

# Part 2

```python
from collections import defaultdict
n_timelines = defaultdict(int)
n_timelines[(0, 70)] += 1
```

```python
beam_positions = {(0, 70)}
while beam_positions:
    new_positions = set()
    for p in beam_positions:
        candidate_position = (p[0] + 1, p[1])
        if aoc.is_outside_bounds(candidate_position, s):
            continue
        if diagram[candidate_position[0]][candidate_position[1]] == "^":
            cp1, cp2 = (p[0] + 1, p[1] - 1), (p[0] + 1, p[1] + 1)
            n_timelines[cp1] += n_timelines[p]
            n_timelines[cp2] += n_timelines[p]
            new_positions.update([cp1, cp2])
        else:
            n_timelines[candidate_position] += n_timelines[p]
            new_positions.update([candidate_position])
    beam_positions = new_positions.copy()

sum(n_timelines[(s[0] - 1, x)] for x in range(s[1] + 1))
```
