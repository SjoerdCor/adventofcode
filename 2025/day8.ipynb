```python
import copy
import numpy as np
import aoc
```

```python
input_ = aoc.read("day8.txt").splitlines()
junctionbox_locations = []
for i in input_:
    p = i.split(',')
    junctionbox_locations.append((int(p[0]), int(p[1]), int(p[2])))
junctionbox_locations.sort()
```

```python
distances = {}
for i, p1 in enumerate(junctionbox_locations):
    for p2 in junctionbox_locations[i+1:]:
        distances[(p1, p2)] = (p2[0] - p1[0])**2 + (p2[1] - p1[1])**2 + (p2[2] - p1[2])**2

distances = [(key, value) for key, 
               value in sorted(distances.items(), 
                               key=lambda item: item[1])]
```

```python
seen = set()
circuits = []
for key, value in distances[:1000]:
    p1, p2 = key
    if p1 not in seen and p2 not in seen:
        circuits.append([p1, p2])
    if p1 in seen and p2 not in seen:
        for circuit in circuits:
            if p1 in circuit:
                circuit.append(p2)
    if p2 in seen and p1 not in seen:
        for circuit in circuits:
            if p2 in circuit:
                circuit.append(p1)
    if p1 in seen and p2 in seen:
        new_circuits = []
        new_circuit = set()
        for i in range(len(circuits)-1, -1, -1):
            if p1 in circuits[i] or p2 in circuits[i]:
                new_circuit.update(circuits.pop(i))
            else: 
                new_circuits.append(circuits[i])
        new_circuits.append(list(new_circuit))
        circuits = copy.deepcopy(new_circuits)
    seen.update([p1])
    seen.update([p2])

lens = [len(c) for c in circuits]
np.prod(sorted(lens, reverse=True)[:3])
```
