```python
with open("../data/day23.txt") as f:
    data = f.read().strip()

data = set([tuple(connection.split("-")) for connection in data.split("\n")])
```

```python
import itertools
```

```python
computers = set()
for row in data:
    computers.add(row[0])
    computers.add(row[1])
# computers
```

```python
len(computers)
```

```python
three_interconnected = set()


def are_connected(comp1, comp2, data=data):
    return ((comp1, comp2) in data) or ((comp2, comp1) in data)


for comp1, comp2, comp3 in itertools.combinations(computers, 3):
    if (
        are_connected(comp1, comp2)
        and are_connected(comp2, comp3)
        and are_connected(comp3, comp1)
    ):
        three_interconnected.add((comp1, comp2, comp3))
```

```python
def has_one_t_computer(comps):
    for comp in comps:
        if comp[0] == "t":
            return True
    return False


sum([has_one_t_computer(comps) for comps in three_interconnected])
```

# Part 2

```python
sorted_computers = sorted(computers)
```

```python
three_interconnected = [tuple(sorted(three)) for three in three_interconnected]
```

```python
connections = {3: three_interconnected}

computer_indexes = {computer: i + 1 for i, computer in enumerate(sorted_computers)}


def try_add_computer(network, sorted_computers=sorted_computers):
    longer_network = set()

    # THis works because it is sorted
    computer_index = computer_indexes[network[-1]]
    for new_computer in sorted_computers[computer_index:]:
        if all(are_connected(new_computer, old_computer) for old_computer in network):
            longer_network.add((*network, new_computer))
    return longer_network


def add_level(interconnected):
    interconnected_one_more = []
    for network in interconnected:
        interconnected_one_more.extend(try_add_computer(network))
    return interconnected_one_more


import tqdm

for i in tqdm.tqdm(range(4, 520)):
    new_networks = add_level(connections[i - 1])
    if not new_networks:
        break
    connections[i] = new_networks
```
