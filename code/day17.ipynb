```python
with open("../data/day17.txt") as f:
    content = f.read()

sections = content.split("\n\n")

register_lines = sections[0].splitlines()
registers = {
    "A": int(register_lines[0].split(":")[1].strip()),
    "B": int(register_lines[1].split(":")[1].strip()),
    "C": int(register_lines[2].split(":")[1].strip()),
}

program = [int(x) for x in sections[1].split(":")[1].strip().split(",")]
```

```python
def get_combo_operand(registers, operator):
    if 1 <= operator <= 3:
        return operator
    if operator == 4:
        return registers["A"]
    if operator == 5:
        return registers["B"]
    if operator == 6:
        return registers["C"]
    if operator == 7:
        raise ValueError("Does not appear in valid program")


def handle_instruction(opcode, operator, registers, instruction_pointer):
    instruction_addition = 2
    output = ""
    if opcode == 0:
        numerator = registers["A"]
        denominator = 2 ** get_combo_operand(registers, operator)
        registers["A"] = numerator // denominator
    if opcode == 1:
        registers["B"] = registers["B"] ^ operator
    if opcode == 2:
        registers["B"] = get_combo_operand(registers, operator) % 8
    if opcode == 3:
        if registers["A"] != 0:
            instruction_pointer = operator
            instruction_addition = 0
    if opcode == 4:
        registers["B"] = registers["B"] ^ registers["C"]
    if opcode == 5:
        output = str(get_combo_operand(registers, operator) % 8)
    if opcode == 6:
        numerator = registers["A"]
        denominator = 2 ** get_combo_operand(registers, operator)
        registers["B"] = numerator // denominator
    if opcode == 7:
        numerator = registers["A"]
        denominator = 2 ** get_combo_operand(registers, operator)
        registers["C"] = numerator // denominator
    instruction_pointer += instruction_addition
    return instruction_pointer, output, registers


def handle_program(registers, program):
    instruction_pointer = 0
    total_output = []
    while instruction_pointer < len(program):
        opcode = program[instruction_pointer]
        operator = program[instruction_pointer + 1]
        print(opcode, operator)
        instruction_pointer, output, registers = handle_instruction(
            opcode, operator, registers, instruction_pointer
        )
        print(instruction_pointer, output, registers)
        if output:
            total_output.append(output)
    return ",".join(total_output)


handle_program(registers, program)
```
