```python
def get_combo_operand(registers, operator):
    if 1 <= operator <= 3:
        return operator
    if operator == 4:
        return registers["A"]
    if operator == 5:
        return registers["B"]
    if operator == 6:
        return registers["C"]
    if operator == 7:
        raise ValueError("Does not appear in valid program")


def handle_instruction(opcode, operator, registers, instruction_pointer):
    instruction_addition = 2
    output = ""
    if opcode == 0:
        numerator = registers["A"]
        denominator = 2 ** get_combo_operand(registers, operator)
        registers["A"] = numerator // denominator
    if opcode == 1:
        registers["B"] = registers["B"] ^ operator
    if opcode == 2:
        registers["B"] = get_combo_operand(registers, operator) % 8
    if opcode == 3:
        if registers["A"] != 0:
            instruction_pointer = operator
            instruction_addition = 0
    if opcode == 4:
        registers["B"] = registers["B"] ^ registers["C"]
    if opcode == 5:
        output = str(get_combo_operand(registers, operator) % 8)
    if opcode == 6:
        numerator = registers["A"]
        denominator = 2 ** get_combo_operand(registers, operator)
        registers["B"] = numerator // denominator
    if opcode == 7:
        numerator = registers["A"]
        denominator = 2 ** get_combo_operand(registers, operator)
        registers["C"] = numerator // denominator
    instruction_pointer += instruction_addition
    return instruction_pointer, output, registers


def handle_program(registers, program):
    instruction_pointer = 0
    total_output = []
    while instruction_pointer < len(program):
        opcode = program[instruction_pointer]
        operator = program[instruction_pointer + 1]
        print(opcode, operator)
        instruction_pointer, output, registers = handle_instruction(
            opcode, operator, registers, instruction_pointer
        )
        print(instruction_pointer, output, registers)
        if output:
            total_output.append(output)
    return ",".join(total_output)


handle_program(registers, program)
```

```python
# This is a very specific solution, since I know the following things about my program
# 1. it starts at the beginning if a != 0 at the end of the program, and otherwise it ends
# 2. the starting values of B & C do not matter - they are overwritten before they are used
# 3. The value of `a` at the end is a_start \\ 8


@functools.cache
def step(a):

    # PROGRAM
    # (2, 4) - B = A % 8
    # (1, 6) - B = B ^ 6
    # (7, 5) - C = A // 2 ** B
    # (4, 4) - B = B ^ C   B=C
    # (1, 7) - B = B ^ 7   B=0
    # (0, 3) - A = A // 8 A = [1, 7]
    # (5, 5) - out B % 8
    # (3, 0) - move to beginning unless A == 0
    b = (a % 8) ^ 6
    c = a // 2**b
    b ^= c
    b ^= 7
    a //= 8
    return a, [b % 8]


def run_program(a):
    output = []
    while a > 0:
        a, ouptut_step = step(a)
        output.extend(ouptut_step)
    return output


def check_options(program, options):
    output = []
    while output != program:
        new_options = []
        for option in options:
            new_options.extend(range(8 * option, 8 * option + 8))

        options = []
        for option in new_options:
            output = run_program(option)
            if output == program[-len(output) :]:
                options.append(option)

            if output == program:
                return option
    return options


check_options(program, [0])
```
