```python
with open("../data/day5.txt") as f:
    data = f.readlines()
rules = data[: data.index("\n")]
updates = data[data.index("\n") + 1 :]
```

```python
# Clean
rules = [tuple(map(int, rule.strip().split("|"))) for rule in rules]
updates = [list(map(int, update.strip().split(","))) for update in updates]
```

# Part 1

```python
def check_update_v_rule(update, rule):
    if rule[0] not in update or rule[1] not in update:
        return True
    return update.index(rule[0]) < update.index(rule[1])


def get_valid_updates(updates, rules):
    valid_updates = []
    for update in updates:
        if all(check_update_v_rule(update, rule) for rule in rules):
            valid_updates.append(update)
    return valid_updates


valid_updates = get_valid_updates(updates, rules)
```

```python
def get_middle_page(update):
    middle_index = len(update) // 2
    return update[middle_index]


sum(get_middle_page(update) for update in valid_updates)
```

# Part 2

```python
def get_invalid_updates(updates, rules):
    invalid_updates = []
    for update in updates:
        if not all(check_update_v_rule(update, rule) for rule in rules):
            invalid_updates.append(update)
    return invalid_updates
```

```python
def get_invalid_rules(update, rules):
    invalid_rules = []

    for rule in rules:
        if not check_update_v_rule(update, rule):
            invalid_rules.append(rule)
    return invalid_rules
```

```python
def fix_incorrect_rule(update, rule):
    temp_update = update.copy()
    last_index = update.index(rule[1])
    temp_update.remove(rule[0])
    temp_update.insert(last_index, rule[0])
    return temp_update
```

```python
def fix_update(update, rules):
    temp_update = update.copy()
    invalid_rules = get_invalid_rules(temp_update, rules)
    i = 0
    while True:
        # for invalid_rule in invalid_rules:
        invalid_rule = invalid_rules[0]
        temp_update = fix_incorrect_rule(temp_update, invalid_rule)
        invalid_rules = get_invalid_rules(temp_update, rules)

        if not invalid_rules:
            break
        i += 1
        if i == 1_000:
            print(temp_update)
        if i == 10_000:
            print(temp_update, invalid_rules)
            break
    return temp_update
```

```python
invalid_updates = get_invalid_updates(updates, rules)
fixed_updates = []
for invalid_update in invalid_updates:
    fixed_update = fix_update(invalid_update, rules)
    fixed_updates.append(fixed_update)
```

```python
sum(get_middle_page(update) for update in fixed_updates)
```
