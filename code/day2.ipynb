# Part 1
The engineers are trying to figure out which reports are safe. The Red-Nosed reactor safety systems can only tolerate levels that are either gradually increasing or gradually decreasing. So, a report only counts as safe if both of the following are true:

The levels are either all increasing or all decreasing.
Any two adjacent levels differ by at least one and at most three.
In the example above, the reports can be found safe or unsafe by checking those rules:

7 6 4 2 1: Safe because the levels are all decreasing by 1 or 2.
1 2 7 8 9: Unsafe because 2 7 is an increase of 5.
9 7 6 2 1: Unsafe because 6 2 is a decrease of 4.
1 3 2 4 5: Unsafe because 1 3 is increasing but 3 2 is decreasing.
8 6 4 4 1: Unsafe because 4 4 is neither an increase or a decrease.
1 3 6 7 9: Safe because the levels are all increasing by 1, 2, or 3.
So, in this example, 2 reports are safe.

```python
import pandas as pd
```

```python
with open("../data/day2.csv") as f:
    data = f.readlines()
data = [list(map(int, report.strip().split())) for report in data]  # Clean
```

```python
def check_safe(report):
    increasing = report[1] > report[0]
    for frst, nxt in zip(report[:-1], report[1:]):
        if increasing and (1 <= (nxt - frst) <= 3):
            continue
        elif not increasing and (1 <= (frst - nxt) <= 3):
            continue
        else:
            return False
    return True


safety = [check_safe(level) for level in data]
print(sum(safety))
```

# Part 2 

The Problem Dampener is a reactor-mounted module that lets the reactor safety systems tolerate a single bad level in what would otherwise be a safe report. It's like the bad level never happened!

```python
def check_safe_with_missing(report):
    safe = check_safe(report)
    if safe:
        return True
    for i_removed in range(len(report)):
        report_with_level_removed = [
            report[i] for i in range(len(report)) if i != i_removed
        ]
        if check_safe(report_with_level_removed):
            return True
    return False


safety = [check_safe_with_missing(report) for report in data]
print(sum(safety))
```
