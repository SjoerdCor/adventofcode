```python
import os
import numpy as np


def file_to_array(filename):
    loc = os.path.join("..", "data", filename)
    with open(loc) as f:
        data = f.read()
    return np.array([list(map(int, line)) for line in data.split("\n")])


data = file_to_array("day10.txt")
```

```python
def check_outside_bounds(test_position, size):
    return (
        test_position[0] < 0
        or test_position[1] < 0
        or test_position[0] >= size[0]
        or test_position[1] >= size[1]
    )


def is_valid_step(from_position, to_position, data):
    outside = check_outside_bounds(to_position, data.shape)
    if outside:
        return False
    return (data[*to_position] - data[*from_position]) == 1
```

```python
trailheads = np.argwhere(data == 0)
scores = []
for trailhead in trailheads:
    current_positions = [trailhead]
    peaks_reached = []
    next_steps = [
        np.array((-1, 0)),
        np.array((0, -1)),
        np.array((1, 0)),
        np.array((0, 1)),
    ]

    while current_positions:
        current_position = current_positions[0]
        if data[*current_position] == 9:
            peaks_reached.append(tuple(current_position))
            del current_positions[0]
            continue
        for possible_step in next_steps:

            test_position = current_position + possible_step
            valid_step = is_valid_step(current_position, test_position, data)
            if valid_step:
                current_positions.append(test_position)

        del current_positions[0]
    score = len(set(peaks_reached))
    scores.append(score)
sum(scores)
```

```python
trailheads = np.argwhere(data == 0)
scores = []
distinct_route_scores = []
for trailhead in trailheads:
    current_positions = [trailhead]
    trailhead_score = 0
    peaks_reached = []
    next_steps = [
        np.array((-1, 0)),
        np.array((0, -1)),
        np.array((1, 0)),
        np.array((0, 1)),
    ]

    while current_positions:
        current_position = current_positions[0]
        if data[*current_position] == 9:
            trailhead_score += 1
            peaks_reached.append(tuple(current_position))
            del current_positions[0]
            continue
        for possible_step in next_steps:

            test_position = current_position + possible_step
            valid_step = is_valid_step(current_position, test_position, data)
            if valid_step:
                current_positions.append(test_position)

        del current_positions[0]
    score = len(set(peaks_reached))
    scores.append(score)
    distinct_route_scores.append(trailhead_score)
sum(scores)
```
