```python
import pandas as pd
```

```python
s1 = pd.Series(array1)
s2 = pd.Series(array2)
```

```python
occ1 = s1.value_counts().rename("OccurrenceArray1")
occ2 = s2.value_counts().rename("OccurrenceArray2")
```

```python
occ1.to_frame().join(occ2).fillna(0).sort_values("OccurrenceArray2")
```

```python
(
    occ1.to_frame()
    .join(occ2)
    .fillna(0)
    .assign(
        TotalOccurrence=lambda df: df["OccurrenceArray1"].mul(df["OccurrenceArray2"])
    )
    .reset_index()
    .assign(SimilarityScore=lambda df: df["index"] * df["TotalOccurrence"])[
        "SimilarityScore"
    ]
    .astype(int)
    .sum()
)
```
