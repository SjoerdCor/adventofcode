```python
with open("../data/day7.txt") as f:
    data = f.readlines()
data = [line.strip().split(":") for line in data]

result = []
for line in data:
    test_value = int(line[0])
    numbers = [int(x) for x in line[1].split()]
    result.append((test_value, numbers))

data = result
```

# Part 1

```python
import itertools
```

```python
def get_equation_outcome(numbers, operators):
    result = numbers.pop(0)
    for number, operator in zip(numbers, operators):
        if operator == "*":
            result *= number
        elif operator == "+":
            result += number
        else:
            raise ValueError(f"{operator=}")
    return result


def check_test_value_against_numbers(test_value, numbers):
    operator_options = itertools.product(["*", "+"], repeat=len(numbers) - 1)
    for operators in operator_options:
        number_try = numbers.copy()
        outcome = get_equation_outcome(number_try, operators)
        if outcome == test_value:
            return True
    return False
```

```python
valid_testdata = []
for test_value, numbers in data:
    if check_test_value_against_numbers(test_value, numbers):
        valid_testdata.append(test_value)
```

```python
sum(valid_testdata)
```

# Part 2

```python
def get_equation_outcome(numbers, operators):
    result = numbers.pop(0)
    for number, operator in zip(numbers, operators):
        if operator == "*":
            result *= number
        elif operator == "+":
            result += number
        elif operator == "|":
            result = int(str(result) + str(number))
        else:
            raise ValueError(f"{operator=}")
    return result


def check_test_value_against_numbers(test_value, numbers):
    operator_options = itertools.product(["*", "+", "|"], repeat=len(numbers) - 1)
    for operators in operator_options:
        number_try = numbers.copy()
        outcome = get_equation_outcome(number_try, operators)
        if outcome == test_value:
            return True
    return False
```

```python
valid_testdata = []
for test_value, numbers in data:
    if check_test_value_against_numbers(test_value, numbers):
        valid_testdata.append(test_value)
sum(valid_testdata)
```
