```python
import tqdm
import functools
```

```python
with open("../data/day19.txt") as f:
    data = f.read()
```

```python
sections = data.split("\n\n")
available_patterns = frozenset(sections[0].split(", "))
desired_designs = sections[1].split("\n")
```

```python
@functools.cache
def get_possible_patterns(design, available_patterns):
    possible = []
    for pattern in available_patterns:
        if design[: len(pattern)] == pattern:
            possible.append(pattern)
    return possible


@functools.cache
def can_make_design(design, available_patterns=available_patterns):
    possible_patterns_first = get_possible_patterns(design, available_patterns)
    if not possible_patterns_first:
        return False

    possible = []
    for used_pattern in possible_patterns_first:
        if used_pattern == design:
            return True
        remaining = design[len(used_pattern) :]
        possible.append(can_make_design(remaining))
    return any(possible)


sum(can_make_design(design) for design in tqdm.tqdm(desired_designs))
```

# Part 2

```python
@functools.cache
def number_of_possible_designs(design, available_patterns=available_patterns):
    possible_patterns_first = get_possible_patterns(design, available_patterns)
    if not possible_patterns_first:
        return 0

    possible = []

    for used_pattern in possible_patterns_first:
        if used_pattern == design:
            possible.append(1)
        remaining = design[len(used_pattern) :]
        possible.append(number_of_possible_designs(remaining))
    return sum(possible)


sum(number_of_possible_designs(design) for design in tqdm.tqdm(desired_designs))
```
