```python
test_input = "125 17"
actual_input = "92 0 286041 8034 34394 795 8 2051489"
data = list(map(int, actual_input.split()))
```

```python
import functools
```

```python
@functools.cache
def evolve_stone(number):

    if number == 0:

        return [1]

    string_num = str(number)
    n_digits = len(string_num)

    if (n_digits % 2) == 0:
        return [int(string_num[: n_digits // 2]), int(string_num[n_digits // 2 :])]

    return [number * 2024]


def evolve_line(line):

    new_line = []
    for stone in line:

        new_line.extend(evolve_stone(stone))

    return new_line
```

```python
n_blinks = 25
line = data
for _ in range(n_blinks):
    line = evolve_line(line)
```

# Part 2
The number of stones grows exponentially and is completely infeasible to brute force. Luckily, the stones evolve independently so we can create an iterative function which is instantaneous

```python
@functools.cache
def n_stones_after_n_steps(number, n_steps):
    next_step = evolve_stone(number)
    if n_steps == 1:
        return len(next_step)
    return sum(n_stones_after_n_steps(stone, n_steps - 1) for stone in next_step)
```

```python
sum(n_stones_after_n_steps(stone, 75) for stone in data)
```
