```python
with open("../data/day14.txt") as f:
    data = f.readlines()
```

```python
import numpy as np
```

```python
parsed_data = [
    (
        np.array(
            tuple(map(int, item.split()[0][2:].split(",")))
        ),  # Extract and parse "p"
        np.array(
            tuple(map(int, item.split()[1][2:].split(",")))
        ),  # Extract and parse "v"
    )
    for item in data
]
```

```python
def calculate_position(starting_position, velocity, n_steps=100, width=101, height=103):
    final_position = starting_position + velocity * n_steps
    return (final_position[1] % height, final_position[0] % width)


positions = [calculate_position(*robot) for robot in parsed_data]
```

```python
def get_quadrant(position, height, width):
    up = position[0] < (height // 2)
    down = position[0] > (height // 2)
    left = position[1] < (width // 2)
    right = position[1] > (width // 2)
    if up and left:
        return 1
    if up and right:
        return 2
    if down and left:
        return 3
    if down and right:
        return 4
    return None
```

```python
quadrants = [get_quadrant(position, width=101, height=103) for position in positions]
quadrants = {[quadrant for quadrant in quadrants if quadrant is not None]}
```

```python
from collections import Counter

counter = Counter(quadrants)
```

```python
import math

math.prod(counter.values())
```

```python
def display_positions(positions, width=101, height=103):
    arr = np.array([["."] * height] * width)
    for x, y in positions:
        arr[y, x] = "*"
    result = ""
    for row in arr:
        result += "".join(row) + "\n"
    if "*" * 5 in result:
        print(result)


def calculate_position(starting_position, velocity, n_steps=100, width=101, height=103):
    final_position = starting_position + velocity * n_steps
    return (final_position[1] % height, final_position[0] % width)


for steps in range(100):
    positions = [calculate_position(*robot, n_steps=steps) for robot in parsed_data]
    print(steps)
    display_positions(positions)
    print("-" * 200)
```
