```python
with open("../data/day14.txt") as f:
    data = f.readlines()
```

```python
import numpy as np
from collections import Counter
import math
```

```python
parsed_data = [
    (
        np.array(
            tuple(map(int, item.split()[0][2:].split(",")))
        ),  # Extract and parse "p"
        np.array(
            tuple(map(int, item.split()[1][2:].split(",")))
        ),  # Extract and parse "v"
    )
    for item in data
]
```

```python
def calculate_position(starting_position, velocity, n_steps=100, width=101, height=103):
    final_position = starting_position + velocity * n_steps
    return (final_position[0] % width, final_position[1] % height)


positions_new = [calculate_position(*robot) for robot in parsed_data]
```

```python
def display_positions(positions, width=101, height=103, show=True):
    arr = np.array([["."] * width] * height)
    for x, y in positions:
        arr[y, x] = "*"
    result = ""
    for row in arr:
        result += "".join(row) + "\n"
    if show:
        print(result)
    else:
        return result


display_positions(positions_new)
```

```python
def get_quadrant(position, height=103, width=101):
    up = position[1] < (height // 2)
    down = position[1] > (height // 2)
    left = position[0] < (width // 2)
    right = position[0] > (width // 2)
    if up and left:
        return 1
    if up and right:
        return 2
    if down and left:
        return 3
    if down and right:
        return 4
    return None
```

```python
quadrants = [get_quadrant(position) for position in positions_new]
```

```python
counter = Counter(quadrants)
counter
```

```python
math.prod([v for k, v in counter.items() if k is not None])
```

```python
for steps in range(10_000):
    positions = [calculate_position(*robot, n_steps=steps) for robot in parsed_data]
    result = display_positions(positions, show=False)
    if "*" * 10 in result:
        print(steps)
        print(result)
        print("-" * 200)
```
